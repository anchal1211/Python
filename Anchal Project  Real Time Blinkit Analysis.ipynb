```python
print("Data Cleaning")
```

```python
# Fill missing 'Item Weight' with the mean value of the column
df['Item Weight'] = df['Item Weight'].fillna(df['Item Weight'].mean())
```

```python
# Verify that there are no null values in the column
print("Null values in 'Item Weight' after filling:\n", df['Item Weight'].isnull().sum())
```

```python
# Check for null values
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print("\nNumber of Duplicate Rows:", df.duplicated().sum())
```

```python
# Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows before dropping: {duplicate_count}")
```

```python
# Drop duplicate rows if any exist
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates dropped.")
else:
    print("No duplicates found.")
```

```python
# Verify the result
print(f"Number of duplicate rows after dropping: {df.duplicated().sum()}")
print(f"Shape of the dataset after removing duplicates: {df.shape}")
```

```python
# Verify the cleaning process
print("Null values after cleaning:\n", df.isnull().sum())
print("\nShape after removing duplicates:", df.shape)
```

```python
# Drop unnecessary or unwanted columns (e.g., Dropana)
# If you have a column named 'Dropana' or any other column to drop, you can use:
df = df.drop(columns=['Dropana'], errors='ignore')  # errors='ignore' avoids errors if column is missing
```

```python
# Sorting by 'Sales' column
df = df.sort_values(by='Sales', ascending=False)  # Sorting in descending order to see highest sales first
print("\nSorted Data by Sales (Top 5 rows):")
print(df.head())
```

```python
# Handle negative values (if any)
# Check for negative values in numeric columns like 'Sales' or 'Rating'

# Check negative values in 'Sales'
negative_sales_count = (df['Sales'] < 0).sum()
print(f"Negative Sales count: {negative_sales_count}")

# Check negative values in 'Rating'
negative_rating_count = (df['Rating'] < 0).sum()
print(f"Negative Rating count: {negative_rating_count}")
```

```python
# Check if the 'Sales' and 'Rating' columns are numeric
print(f"Sales column dtype: {df['Sales'].dtype}")
print(f"Rating column dtype: {df['Rating'].dtype}")
```

```python
# Handle negative values: Replace negative sales and ratings with 0
if negative_sales_count > 0:
    print(f"Negative Sales found: {negative_sales_count}")
    df.loc[df['Sales'] < 0, 'Sales'] = 0  # Replace negative sales with 0
else:
    print("No negative values found in 'Sales'.")

if negative_rating_count > 0:
    print(f"Negative Ratings found: {negative_rating_count}")
    df.loc[df['Rating'] < 0, 'Rating'] = 0  # Replace negative ratings with 0
else:
    print("No negative values found in 'Rating'.")
```

```python
# Check the result after handling negative values
print("\nCheck final values:")
print(f"Negative Sales count after handling: {(df['Sales'] < 0).sum()}")
print(f"Negative Rating count after handling: {(df['Rating'] < 0).sum()}")
```

```python
# Verify the cleaning process
print("Null values after cleaning:\n", df.isnull().sum())
print("\nShape after removing duplicates:", df.shape)
```

```python
# Step 5: Final check after cleaning, sorting, and handling negative values
print("\nFinal Check:")
print("\nFinal Check:")
print(f"Shape of the cleaned dataset: {df.shape}")
print(f"Null values after cleaning: {df.isnull().sum()}")
print(f"Number of duplicate rows after cleaning: {df.duplicated().sum()}")
print(f"Number of negative sales entries fixed: {negative_sales_count}")
print(f"Number of negative rating entries fixed: {negative_rating_count}")
```

```python
# Data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [12, 15, 7, 10, 5]

# Set the style and background color (replace 'seaborn-darkgrid' with a valid style)
plt.style.use('ggplot')  # Apply a predefined style that is likely available
plt.figure(figsize=(10, 6), facecolor='#f0f0f0')  # Set background color for the figure

# Create the bar chart
plt.bar(categories, values, color='skyblue')

# Set chart title and labels
plt.title('Bar Chart Example', fontsize=16, color='darkblue')
plt.xlabel('Categories', fontsize=12, color='darkgreen')
plt.ylabel('Values', fontsize=12, color='darkgreen')

# Show the plot
plt.show()
```

```python
import matplotlib.colors as mcolors

# Data for the bar graph
categories = ['A', 'B', 'C', 'D', 'E']
values = [12, 15, 7, 10, 5]

# Set the figure background color to black
plt.figure(figsize=(10, 6), facecolor='black')

# Define a colormap with different colors
colors = list(mcolors.TABLEAU_COLORS.values())  # You can use other color palettes here

# Create the bar graph with shiny effect
bars = plt.bar(categories, values, color=colors)

# Apply a shiny effect (lighter edge color with a glow-like effect)
for bar in bars:
    bar.set_edgecolor('white')  # Set edge color to white for a glow effect
    bar.set_linewidth(1.5)  # Set edge line width

# Set chart title and labels with white color
plt.title('Bar Graph with Shiny Colors and Black Background', fontsize=16, color='white')
plt.xlabel('Categories', fontsize=12, color='white')
plt.ylabel('Values', fontsize=12, color='white')

# Change axis tick colors to white
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')

# Show the plot
plt.show()
```

```python
# Data for pie chart
sizes = [30, 20, 25, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

# Set the style and background color
plt.style.use('ggplot')  # Apply a predefined style
plt.figure(figsize=(8, 8), facecolor='#ffffff')  # Set background color for the figure

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Set chart title
plt.title('Pie Chart Example', fontsize=16, color='darkblue')

# Show the plot
plt.show()
```

```python
df['Outlet Size'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'orange', 'lightgreen'], figsize=(7, 7))
plt.title('Distribution of Outlet Sizes')
plt.ylabel('')
plt.show()
```

```python
# Data for line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set the style and background color
plt.style.use('bmh')  # Apply a predefined style
plt.figure(figsize=(10, 6), facecolor='#f9f9f9')  # Set background color for the figure

# Create the line chart
plt.plot(x, y, label='Sine Wave', color='orange', linewidth=2)

# Set chart title and labels
plt.title('Line Chart Example', fontsize=16, color='darkblue')
plt.xlabel('X values', fontsize=12, color='darkgreen')
plt.ylabel('Y values', fontsize=12, color='darkgreen')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
